```python
bin_no = 12

# Function to calculate Symbol / Bin_Number
def which_bin(val, min, max):
    L = min - 1
    for i in range (1, bin_no):
        R = min + (max - min) / bin_no * i
        if  L < val <= R:
            return i

        L = R

    return bin_no
```

```python
# Assign Symbol / Bin_number to each return value
min1 = df1['Log Return(Rt)'].min() 
max1 = df1['Log Return(Rt)'].max()

min2 = df2['Log Return(Rt)'].min() 
max2 = df2['Log Return(Rt)'].max()

df1['Symbol_x'] = df1['Log Return(Rt)'].apply(lambda value : which_bin(value, min1, max1)) # apply function to each element of the column
df2['Symbol_y'] = df2['Log Return(Rt)'].apply(lambda value : which_bin(value, min2, max2))
```

```python
# Merge two dataframe based on the same date
df = pd.merge(df1, df2, on='Date', how='inner')

df
```

```python
from collections import Counter

# Initialize counters
Fxy_counter = Counter((row.Symbol_x, row.Symbol_y) for row in df.itertuples())
Fx_counter = Counter(row.Symbol_x for row in df.itertuples())
Fy_counter = Counter(row.Symbol_y for row in df.itertuples())
n=bin_no+1
# Convert to 2D and 1D arrays
Fxy = [[Fxy_counter.get((x, y), 0) for y in range(n)] for x in range(n)]
Fx = [Fx_counter.get(x, 0) for x in range(n)]
Fy = [Fy_counter.get(y, 0) for y in range(n)]
```

```python
# Total frequency
total_frequency = sum(Fx)

# Calculate probabilities
Pxy = [[Fxy[x][y] / total_frequency if total_frequency > 0 else 0 for y in range(n)] for x in range(n)]
Px = [Fx[x] / total_frequency if total_frequency > 0 else 0 for x in range(n)]
Py = [Fy[y] / total_frequency if total_frequency > 0 else 0 for y in range(n)]
```

```python
import math

# Calculate mutual information Ixy using a generator expression
Ixy = sum(
    Pxy[i][j] * math.log2(Pxy[i][j] / (Px[i] * Py[j])) 
    for i in range(1, n) 
    for j in range(1, n) 
    if Pxy[i][j] > 0 and Px[i] * Py[j] > 0
)

```
